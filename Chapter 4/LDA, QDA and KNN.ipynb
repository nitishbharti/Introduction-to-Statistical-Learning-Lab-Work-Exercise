# Linear Discriminant Analysis

```r
library('MASS')
library('ISLR')
```

```r
training_data = Smarket[Smarket$Year<2005,]
test_data = Smarket[Smarket$Year>=2005,]
```

```r
lda.fit = lda(Direction~Lag1+Lag2,data=Smarket)
lda.fit
```

```r
plot(lda.fit)
```

```r
lda.pred = predict(lda.fit,test_data)
lda.pred
```

```r
plot(training_data$Lag1,training_data$Lag2)
abline(lda.fit, col="red", lwd=3)
```

# Confusion Matrix

```r
table(lda.pred$class,test_data$Direction)
```

```r
mean(lda.pred$class!=test_data$Direction)
```

### Error is 44% slightly better than Logistic Regression.

# Quadratic Discriminant Analysis

```r
qda.fit = qda(Direction~Lag1+Lag2, data=Smarket)
qda.fit
```

```r
qda.pred = predict(qda.fit,test_data)
qda.pred
```

```r
table(qda.pred$class, test_data$Direction)
```

```r
mean(qda.pred$class!=test_data$Direction)
```

### Test Error is 43% slightly better than LDA.

# K- Nearest Neighbour

```r
library('class')
```

```r
train_x = cbind(training_data$Lag1,training_data$Lag2)
test_x = cbind(test_data$Lag1,test_data$Lag2)
train_direction = training_data$Direction
test_direction = test_data$Direction
```

```r
set.seed(1)
knn.pred = knn(train_x,test_x,train_direction,k=3)
```

```r
table(knn.pred,test_direction)
mean(knn.pred!=test_direction)
```

# Test error is 46%.

# On above dataset QDA is perfoming best !
