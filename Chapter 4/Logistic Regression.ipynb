# Logistic Regression

```r
glm.fits = glm(Direction~Lag1+Lag2+Lag3+Lag4+Lag5+Volume, data=Smarket, family=binomial)
summary(glm.fits)
```

```r
# Tells about dummy variable created by R for response.
contrasts(Smarket$Direction)
```

```r
glm.probs = predict(glm.fits, type="response") # Type response tells R to send response of type Pr(Y=1|X)
glm.probs[1:10]
```

```r
glm.pred = rep("Down",1250);
glm.pred[glm.probs>0.5] = "Up"
```

# Confusion Matrix

```r
table(glm.pred,Smarket$Direction)
```

### Let's split the data into training and test and calculate % test error

```r
training_data = Smarket[Smarket$Year<2005,]
test_data = Smarket[Smarket$Year>=2005,]

glm.fit = glm(Direction~Lag1+Lag2+Lag3+Lag4+Lag5+Volume, data=training_data, family="binomial")
summary(glm.fit)
```

```r
glm.probs = predict(glm.fit,test_data,type="response")
dim(test_data)

glm.pred = rep("Down",252)
glm.pred[glm.probs>0.5] = "Up"
```

```r
table(glm.pred,test_data$Direction)
```

```r
mean(glm.pred!=test_data$Direction)
```

### We got test error as 52%, which is even worse than flipping a fair coin. But this result is not surprising as we can't predict stock market behavior based only upon previous lags.
