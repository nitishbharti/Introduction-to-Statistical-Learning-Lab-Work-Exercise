# Graphics

```r
x = rnorm(100)
y = rnorm(100)
plot(x,y,xlab='X Axis',ylab='Y Axis')
```

```r
pdf = ("Figure.pdf")
plot(x,y,col="green")
dev.off()
```

```r
x = seq(-pi,pi,length=50)
y=x
f = outer(x,y,function(x,y) cos(y)/1+x^2)
contour(x,y,f)
```

```r
image(x,y,f)
```

```r
persp(x,y,f)
persp(x,y,f,theta =30,phi =70)   # Angle at which the plot is viewed
```

# Additional Graphical and numerical summaries

```r
plot(Auto$cylinders,Auto$mpg)
```

### cylinder variable is stored as a numeric vector, so R treated this as quantitative, but one may prefer to use as qualitative

```r
plot(as.factor(Auto$cylinders),Auto$mpg, col="green", xlab='Cylinders', ylab= 'MPG')
```

```r
hist(Auto$mpg, col="green")
```

```r
pairs(~ mpg + displacement + horsepower + weight +
acceleration , Auto)
```
